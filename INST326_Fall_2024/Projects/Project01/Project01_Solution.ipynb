#### 1. Personal Finance Tracker
>  Create a simple personal finance tracker to help users manage their income and expenses.

**Key Topics Covered:**<br>
Variables and Operators<br>
Data Types<br>
Input and Output<br>
Conditionals<br>
Loops<br>
Functions<br>
Collections<br>
**Steps:**<br>
**Welcome Message:** Print a welcome message and explain the purpose of the program.<br>
**Input Income and Expenses:** Use <font color=green>input()</font> to take the user's income and a list of expenses.<br>
**Calculate Total Expenses:** Use a loop to calculate the total expenses.<br>
**Calculate Savings:** Subtract total expenses from income.<br>
**Display Results:** Print the total income, total expenses, and remaining savings.<br>
**Save to File:** Save the details to a text file for future reference.<br>

```python
def get_expenses():
    expenses = []
    while True:
        expense = input("Enter an expense (or 'done' to finish): ")
        if expense.lower() == 'done':
            break
        try:
            expenses.append(float(expense))
        except ValueError:
            print("Please enter a valid number.")
    return expenses

def main():
    print("Welcome to the Personal Finance Tracker")
    income = float(input("Enter your total income: "))

    print("Enter your expenses:")
    expenses = get_expenses()

    total_expenses = sum(expenses)
    savings = income - total_expenses

    print(f"\nTotal Income: ${income}")
    print(f"Total Expenses: ${total_expenses}")
    print(f"Savings: ${savings}")

    with open("finance_report.txt", "w") as file:
        file.write(f"Income: ${income}\n")
        file.write(f"Expenses: {expenses}\n")
        file.write(f"Total Expenses: ${total_expenses}\n")
        file.write(f"Savings: ${savings}\n")

if __name__ == "__main__":
    main()
```

#### 2. Simple Quiz Game
>  Create a quiz game that asks the user a series of questions and keeps track of their score.

**Key Topics Covered:**<br>
Variables and Operators<br>
Data Types<br>
Input and Output<br>
Conditionals<br>
Loops<br>
Functions<br>
Collections<br>
**Steps:**<br>
**Welcome Message:** Print a welcome message and explain the rules of the game.<br>
**Define Questions:** Use a list of dictionaries to store questions and answers.<br>
**Ask Questions:** Loop through the questions, get the user's answer, and check if it's correct.<br>
**Keep Score:** Maintain a score counter for correct answers.<br>
**Display Results:** Print the user's final score and a thank you message.<br>

```python
# Solution - enter your code solution below
def ask_question(question, answer):
    user_answer = input(question + " ")
    return user_answer.strip().lower() == answer.lower()

def main():
    print("Welcome to the Simple Quiz Game")
    questions = [
        {"question": "What is the capital of France?", "answer": "Paris"},
        {"question": "What is 2 + 2?", "answer": "4"},
        {"question": "What is the color of the sky on a clear day?", "answer": "blue"}
    ]

    score = 0
    for q in questions:
        if ask_question(q["question"], q["answer"]):
            score += 1

    print(f"\nYour final score is {score}/{len(questions)}")
    print("Thank you for playing!")

if __name__ == "__main__":
    main()
```

#### 3. Contact Book
>  Create a contact book application that allows users to add, search, and delete contacts.

**Key Topics Covered:**<br>
Variables and Operators<br>
Data Types<br>
Input and Output<br>
Conditionals<br>
Loops<br>
Functions<br>
Collections<br>
File Handling<br>
**Steps:**<br>
**Welcome Message:** Print a welcome message and provide a menu of options.<br>
**Add Contact:** Allow the user to add a new contact with name and phone number.<br>
**Search Contact:** Allow the user to search for a contact by name.<br>
**Delete Contact:** Allow the user to delete a contact by name.<br>
**Save to File:** Save the contacts to a text file and load them at the start.<br>


```python
def load_contacts():
    try:
        with open("contacts.txt", "r") as file:
            contacts = [line.strip().split(",") for line in file.readlines()]
        return {name: phone for name, phone in contacts}
    except FileNotFoundError:
        return {}

def save_contacts(contacts):
    with open("contacts.txt", "w") as file:
        for name, phone in contacts.items():
            file.write(f"{name},{phone}\n")

def add_contact(contacts):
    name = input("Enter contact name: ")
    phone = input("Enter contact phone number: ")
    contacts[name] = phone

def search_contact(contacts):
    name = input("Enter the name to search: ")
    phone = contacts.get(name)
    if phone:
        print(f"Name: {name}, Phone: {phone}")
    else:
        print("Contact not found.")

def delete_contact(contacts):
    name = input("Enter the name to delete: ")
    if name in contacts:
        del contacts[name]
        print("Contact deleted.")
    else:
        print("Contact not found.")

def main():
    print("Welcome to the Contact Book")
    contacts = load_contacts()

    while True:
        print("\nMenu:")
        print("1. Add Contact")
        print("2. Search Contact")
        print("3. Delete Contact")
        print("4. Exit")

        choice = input("Choose an option: ")
        if choice == "1":
            add_contact(contacts)
        elif choice == "2":
            search_contact(contacts)
        elif choice == "3":
            delete_contact(contacts)
        elif choice == "4":
            save_contacts(contacts)
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
