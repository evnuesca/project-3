#### Option 1: A Calculator App.

```python
import tkinter as tk

class Calculator:
    def __init__(self):
        self.current = ""
        self.total = 0
        self.operator = ""
    
    def add_digit(self, digit):
        self.current += str(digit)
        self.update_display()
    
    def set_operator(self, operator):
        if self.current:
            self.total = int(self.current)
            self.current = ""
            self.operator = operator
    
    def calculate(self):
        if self.current and self.operator:
            if self.operator == "+":
                self.total += int(self.current)
            elif self.operator == "-":
                self.total -= int(self.current)
            elif self.operator == "*":
                self.total *= int(self.current)
            elif self.operator == "/":
                self.total /= int(self.current)
            self.current = str(self.total)
            self.update_display()
            self.current = ""
    
    def update_display(self):
        display_var.set(self.current or self.total)
    
    def clear(self):
        self.current = ""
        self.total = 0
        self.operator = ""
        self.update_display()

app = tk.Tk()
app.title("Simple Calculator")

calc = Calculator()
display_var = tk.StringVar()

display = tk.Entry(app, textvariable=display_var, font=("Arial", 20), bd=10, insertwidth=4, borderwidth=4)
display.grid(row=0, column=0, columnspan=4)

buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
    ('0', 4, 0), ('C', 4, 1), ('=', 4, 2), ('+', 4, 3),
]

for (text, row, col) in buttons:
    action = lambda x=text: calc.add_digit(x) if x.isdigit() else \
             (calc.set_operator(x) if x in "+-*/" else calc.calculate() if x == "=" else calc.clear())
    tk.Button(app, text=text, padx=20, pady=20, font=("Arial", 18), command=action).grid(row=row, column=col)

app.mainloop()
```
