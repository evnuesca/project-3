#### 5. Label Update
>  Create a Tkinter window with an entry widget, a button, and a label. When the button is clicked, update the label to display the content of the entry widget.

```python
# Solution - enter your code solution below
import tkinter as tk

def update_label():
    label.config(text=entry.get())

root = tk.Tk()
root.title("Label Update Example")

entry = tk.Entry(root)
entry.pack()

button = tk.Button(root, text="Update Label", command=update_label)
button.pack()

label = tk.Label(root, text="Original Text")
label.pack()

root.mainloop()
```

#### 6. Combining Widgets
>  Create a Tkinter window with a label, an entry widget, and a button. When the button is clicked, change the label text to "Hello, [name]!" where [name] is the text entered in the entry widget.

```python
# Solution - enter your code solution below
import tkinter as tk

def say_hello():
    name = entry.get()
    label.config(text=f"Hello, {name}!")

root = tk.Tk()
root.title("Greeting Example")

label = tk.Label(root, text="Enter your name:")
label.pack()

entry = tk.Entry(root)
entry.pack()

button = tk.Button(root, text="Greet", command=say_hello)
button.pack()

label = tk.Label(root, text="")
label.pack()

root.mainloop()
```

#### 7. Checkbox Widget
>  Create a Tkinter window with two checkboxes labeled "Option 1" and "Option 2". When a checkbox is checked or unchecked, print the state of the checkboxes to the console. (refer to the second program from the lecture for syntax on creating checkboxes.)

```python
# Solution - enter your code solution below
import tkinter as tk

def show_checkboxes():
    print(f"Option 1: {var1.get()}")
    print(f"Option 2: {var2.get()}")

root = tk.Tk()
root.title("Checkbox Example")

var1 = tk.IntVar()
var2 = tk.IntVar()

checkbox1 = tk.Checkbutton(root, text="Option 1", variable=var1, command=show_checkboxes)
checkbox1.pack()

checkbox2 = tk.Checkbutton(root, text="Option 2", variable=var2, command=show_checkboxes)
checkbox2.pack()

root.mainloop()
```

#### 8. Text Widget
>  Create a Tkinter window with a text widget and a button. When the button is clicked, print the contents of the text widget to the console.

```python
import tkinter as tk

# Function to print the content of the text widget
def print_text():
    content = text_widget.get("1.0", tk.END)  # Get text from the widget starting at line 1, character 0
    print(content)

# Create the main application window
root = tk.Tk()
root.title("Text Widget Example")

# Create a Text widget
text_widget = tk.Text(root, height=10, width=40)
text_widget.pack()

# Create a button widget
print_button = tk.Button(root, text="Print Text", command=print_text)
print_button.pack()

# Run the Tkinter event loop
root.mainloop()
```

#### 9. and 10. Note App (6pts)
>  This exercise will create a window for creating a note, like in a note-taking app. Create a Tkinter window with three labels, an entry box, a text box, and a button. You may also want to create two frames. The first label is a title label that says "Create your note". The second label says "note title: " and goes with the entry box in the first frame.  The third label says "note text: " and goes with the text widget in the second frame. The button contains the text "submit". When the button is pressed, it prints the note title, note text, and current timestamp to the console.

```python
# Solution - enter your code solution below
import tkinter as tk
from datetime import datetime

# Function to handle the button click and print the note details
def submit_note():
    note_title = entry_title.get()
    note_text = text_note.get("1.0", tk.END)
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    print(f"Note Title: {note_title}")
    print(f"Note Text: {note_text}")
    print(f"Timestamp: {timestamp}")

# Create the main application window
root = tk.Tk()
root.title("Note Creator")

# Create a frame for the title and entry
frame1 = tk.Frame(root)
frame1.pack(pady=10)

# Create a frame for the text widget
frame2 = tk.Frame(root)
frame2.pack(pady=10)

# Create the title label
title_label = tk.Label(root, text="Create your note", font=("Helvetica", 16, "bold"))
title_label.pack(pady=10)

# Create the label and entry for the note title
label_title = tk.Label(frame1, text="Note Title: ", font=("Helvetica", 12))
label_title.pack(side=tk.LEFT)

entry_title = tk.Entry(frame1, width=30)
entry_title.pack(side=tk.LEFT, padx=10)

# Create the label and text widget for the note text
label_text = tk.Label(frame2, text="Note Text: ", font=("Helvetica", 12))
label_text.pack(side=tk.TOP, anchor='w')

text_note = tk.Text(frame2, height=10, width=40)
text_note.pack(pady=5)

# Create the submit button
submit_button = tk.Button(root, text="Submit", command=submit_note)
submit_button.pack(pady=20)

# Run the Tkinter event loop
root.mainloop()
```
